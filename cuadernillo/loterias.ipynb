```python
def conteo_numeros(lista):
    dict = {}
    for tirada in lista:
        for valores in tirada:
            if valores in dict.keys():
                valor = dict[valores]
                dict[valores]=valor + 1
            else:
                dict[valores] = 1
    return dict
```

```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

# Preprocesado y análisis
# ==============================================================================
import statsmodels.api as sm
from scipy import stats

# Configuración matplotlib
# ==============================================================================
plt.style.use('ggplot')

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

```python
valores_numeros = df_final.iloc[-1,:]
```

```python
# Histograma + curva normal teórica
# ==============================================================================

# Valores de la media (mu) y desviación típica (sigma) de los datos
mu, sigma = stats.norm.fit(valores_numeros)
```

```python
mu
```

```python
# Valores teóricos de la normal en el rango observado
x_hat = np.linspace(min(valores_numeros), max(valores_numeros), num=100)
y_hat = stats.norm.pdf(x_hat, mu, sigma)
```
